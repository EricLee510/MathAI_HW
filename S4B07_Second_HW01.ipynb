```python
import cv2
import time
ESC = 23 #Ctrl + w
space = 32

#setting
cap = cv2.VideoCapture(0) 
ratio = cap.get(cv2.CAP_PROP_FRAME_WIDTH) / cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
WIDTH = 600
HEIGHT = int(WIDTH/ratio)
cv2.namedWindow("Me", cv2.WINDOW_NORMAL)
font = cv2.FONT_HERSHEY_SIMPLEX

#value
ret = True
capture = False
number = 4
count = 1
show = False


#action
while ret:  
    #key control
    if cv2.waitKey(1) == space:
        capture = True
        lastTime = time.time()
    elif  cv2.waitKey(1) == ESC:
        break
    
    #taking the frame of my camera 
    ret, frame = cap.read() 
    frame = cv2.resize(frame, (WIDTH, HEIGHT))
    frame = cv2.flip(frame,1)
    
    
    #count number and taking picture
    if capture == True and number <= 0:
        cv2.imshow("Me", frame)
        cv2.imwrite(str(count)+".jpeg", frame)
        capture = False
        number = 4
        count += 1
        show = False       
    elif capture == True and time.time() - lastTime >= 1:
        lastTime = time.time()
        number -= 1
        show = True
  
    #the addon of frame 
    if show == True and number > 0:
        cv2.putText(frame, str(number), (10,100), font, 5, (255,255,255), 5,cv2.LINE_AA)
    elif show == True and number == 0:
        frame.fill(255)
        
    #frame show
    cv2.imshow("Me", frame)  
    
    

    
        
        
cv2.destroyAllWindows() 
cap.release()
```
