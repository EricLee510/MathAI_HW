# 找錢



```python
moneyList = [500,100,50,10,5,1]
money = { }
itMoney = int(input("Q: How much is it?: "))
getMoney = int(input("Q: How much does he give?: "))
give = getMoney - itMoney

for i in range(len(moneyList)):
    money[str(moneyList[i])] = give//moneyList[i]
    give = give % moneyList[i]
    print("Data of %: "+ str(give % moneyList[i]))
print("A: You sure give: " + str(money))
print("A: Thanks for coming")
```
