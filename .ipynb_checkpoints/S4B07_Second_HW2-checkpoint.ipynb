```python
import cv2
import os

#value
space = 32
ESC = 27
PN = 0
i = 0
total = 3
ret = True
capture = False

#cap setting
cap = cv2.VideoCapture(0) 
ratio = cap.get(cv2.CAP_PROP_FRAME_WIDTH) / cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
WIDTH = 600 
HEIGHT = int(WIDTH/ratio)   
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt2.xml") 
cv2.namedWindow("Me", cv2.WINDOW_NORMAL) #Action - open a window(window's name, window's type)
font = cv2.FONT_HERSHEY_SIMPLEX


#os folder 1
DirPATH = os.path.join(f"images")
try:
    os.mkdir(DirPATH)
except FileExistsError:
    print("Directory already exists!")
    pass

#action
while ret:
    #key setting 
    if cv2.waitKey(1) == space:
        capture = True
    elif  cv2.waitKey(1) == ESC:
        break
        
    #get frame
    ret, frame = cap.read() 
    frame = cv2.resize(frame, (WIDTH, HEIGHT)) 
    frame = cv2.flip(frame,1) 
    #change gray frame
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) 
    #select detect tool
    faces = face_cascade.detectMultiScale(gray,1.1,5)
    
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(125,0,200),2)
        cv2.putText(frame, str(PN), (x,y-2), font, 2, (125,0,200), 2,cv2.LINE_AA)
        cv2.imshow("Me", frame)  
        if capture == True and w != 0 and h!= 0:
            cv2.imshow("Me", frame) 
            gray1 = gray[y+2:y+h-1,x+2:x+w-1]  
            if i%5 == 0:
                PN += 1
                cv2.imshow("Me", frame) 
                cv2.imwrite( f"images/p0/{PN}.jpeg", gray1)
            if cv2.waitKey(1) == ESC and PN >= total:
                print("stop")
                capture = False
            i += 1
    cv2.imshow("Me", frame)


cv2.destroyAllWindows() 
cap.release() 
```

```python
import cv2
import os
ESC = 27
n = 1
index = 0
total = 155
font = cv2.FONT_HERSHEY_SIMPLEX

#function
def saveImage(face_image,index):
    filename = f"images/p2/{index}.jpeg"
    cv2.imwrite(filename, face_image)
    #print(filename)
    
#folder 1
DirPATH = os.path.join(f"images")
try:
    os.mkdir(DirPATH)
except FileExistsError:
    print("Directory already exists!")
    pass

#folder2
DirPATH = os.path.join(f"images/p2")
try:
    os.mkdir(DirPATH)
except FileExistsError:
    print("Directory already exists!")
    pass

#cap setting
cap = cv2.VideoCapture(0) 
ratio = cap.get(cv2.CAP_PROP_FRAME_WIDTH) / cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
WIDTH = 600 
HEIGHT = int(WIDTH/ratio)   
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt2.xml") 
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.namedWindow("Me", cv2.WINDOW_NORMAL) #Action - open a window(window's name, window's type)
ret = True

while ret:      
    #get frame
    ret, frame = cap.read() 
    frame = cv2.resize(frame, (WIDTH, HEIGHT)) 
    frame = cv2.flip(frame,1) 
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) 
    faces = face_cascade.detectMultiScale(gray,1.1,5)
    
    for (x,y,w,h) in faces:
          if w != 0 and h != 0:
            cv2.putText(frame,str(index),(10,80), font, 3,(0,0,0),2,cv2.LINE_AA)
            cv2.rectangle(frame,(x,y),(x+w,y+h),(125,0,200),2)
            if n%5 == 0:
                face_image = cv2.resize(frame[y:y+h,x:x+w],(400,400))
                saveImage(face_image,index)
                index +=1
            if index >= total:
                #print(f"save{index}photos")
                n = -1
                break
            n+= 1
    cv2.imshow("frame",frame)
    
    if cv2.waitKey(1) == ESC:
        break



cv2.destroyAllWindows() 
cap.release() 
```
