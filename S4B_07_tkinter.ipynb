# 遊戲計算

```python
import tkinter as tk 
import numpy as np
import time 
import tkinter.messagebox

#normal variable
max1 = 99
min1 = 1
bingo = np.random.randint(min1, max1+1)
player = 1
round1 = 0
AI = False
callback = 1
start = True

#using of update window
def update_window():
    number.pack_forget()
    spnNum.pack_forget()
    btn.pack_forget()
    if AI == True:
        AiS_window()
    else:
        twos_window()   

#run update_number() 2 times (一次係玩家回合，一次係AI回合):
def reset_callback(): #用ok按鈕
    global callback
    callback = 1
    update_number()
def reset_callback2(e): #用keyboard
    global callback
    code = e.keycode
    if(code == 13):
        callback = 1
        update_number()
    else:
        return
    
#更新最新的數字範圍和檢測是否bingo    
def update_number():
    global callback
    global num
    global max1
    global min1
    global number
    global spnNum
    global btn
    global player
    global round1
    global AI
    
    #AI 回合時 告訴玩家AI的數字
    if callback == 2 and AI == True: 
        #now = 50
        now = np.random.randint(min1, max1)
        tk.messagebox.showinfo("tip",f"AI number is {now}")
    else:
        now = num.get()
    
    #更新數字範圍
    if now < min1 or now > max1: #超出範圍時
        tk.messagebox.showwarning("Warning","You cannot do that \n You lose this round!")
    elif (now == min1 or now == max1) and now != bingo:  #剛好是等於範圍兩個數字時且不等於答案時
        tk.messagebox.showinfo("tip","Not this number \n You lose this round!")
    elif now > bingo:  #更新範圍
        max1 = now
    elif now < bingo:
        min1 = now
    elif now == bingo and callback == 1 and AI == True: #AI模式時bingo時 ->玩家bingo時
        endtime = time.time()
        tk.messagebox.showinfo(f"result",f" Bingo! \n\n Winner: player{player-1} \n Round: {round1} \n bingo number: {now} \n time:{round(endtime - starttime, 2)}")
        update_window()
        win.destroy()
        return()
    elif now == bingo and callback == 1 and AI == False: #玩家bingo時
        endtime = time.time()
        tk.messagebox.showinfo(f"result",f" Bingo! \n\n Winner: player{player} \n Round: {round1} \n bingo number: {now} \n time:{round(endtime - starttime, 2)}")
        update_window()
        win.destroy()
        return()
    else: ##AI模式時bingo時 ->AI bingo時
        endtime = time.time()
        tk.messagebox.showinfo(f"result",f" Bingo! \n\n Winner: AI \n Round: {round1} \n bingo number: {now} \n time:{round(endtime - starttime, 2)}")
        update_window()
        win.destroy()
        return()
    
    update_window()
    
    if AI == True and callback < 2: #AI模式時bingo時，run多次此function
        callback += 1
        update_number()
```

# 1V1 遊戲介面

```python
def twos_window():
    global num
    global max1
    global min1
    global number
    global spnNum
    global btn
    global player
    global round1
    global start 
    global starttime
    
    #更新回合和玩家
    if start == True:
        player = 0
        start = False
        starttime = time.time()
    if player == 2:
        player = 1
    else:
        player += 1
        
    if player == 1:
        round1 += 1
    
    #clear 以前的畫面
    twoS.pack_forget()
    AiS.pack_forget()
    
    #增加返遊戲訊
    #玩家、回合、範圍的資訊
    number = tk.Label(win, text = f"  player: {player}  round: {round1}  \n    {min1} ~ {max1}    ", fg = "red", bg = "white", font = ("Times", 30, "bold"),relief = "ridge")
    #輸入
    num = tk.IntVar()
    num.set(min1)
    spnNum = tk.Spinbox(win, from_ = min1, to = max1, textvariable = num,font = ("Times", 20, "bold"))
    #ok按鈕
    btn = tk.Button(win, text = "OK",font = ('Helvetica 18 bold',18,'bold'),bg = "white",fg = "black",anchor = "center",command = reset_callback)
    spnNum.bind("<KeyRelease>", reset_callback2)
    
    number.pack(pady = 50)
    spnNum.pack(pady = 5)
    btn.pack(pady = 5)
```

# 1v AI 遊戲介面

```python
def AiS_window():
    global num
    global max1
    global min1
    global number
    global spnNum
    global btn
    global player
    global round1
    global callback
    global AI 
    global start
    global starttime
    
    #clear以前畫面
    twoS.pack_forget()
    AiS.pack_forget()
    
    #開啟AI模式
    AI = True
    
    #更新回合和玩家，並展示出玩家、回合、範圍的資訊
    if start == True:
        start = False
        starttime = time.time()   
    if player == 2:
        number = tk.Label(win, text = f"  player: AI  round: {round1}  \n    {min1} ~ {max1}    ", fg = "red", bg = "white", font = ("Times", 30, "bold"),relief = "ridge")
        player = 1 
    else:
        number = tk.Label(win, text = f"  player: {player}  round: {round1}  \n    {min1} ~ {max1}    ", fg = "red", bg = "white", font = ("Times", 30, "bold"),relief = "ridge")
        player += 1
    
    if player == 1:
        round1 += 1    

    #輸入   
    num = tk.IntVar()
    num.set(min1)
    spnNum = tk.Spinbox(win, from_ = min1, to = max1, textvariable = num,font = ("Times", 20, "bold"))
    #ok按鈕
    btn = tk.Button(win, text = "OK",font = ('Helvetica 18 bold',18,'bold'),bg = "white",fg = "black",anchor = "center",command = reset_callback)
    spnNum.bind("<KeyRelease>", reset_callback2)
    
    number.pack(pady = 50)
    spnNum.pack(pady = 5)
    btn.pack(pady = 5)
```

# 起始介面

```python
#基本設定
win = tk.Tk()
win.geometry("520x450")
win.configure(bg = "black")
win.resizable(False, False)
win.title("Game")

#標題
title = tk.Label(win, text = '開 口 中', font = ('Helvetica 18 bold',30,'bold'),bg = "black",fg = "red",anchor = "center")
title.pack(fill = "x",anchor = "center")

#開始兩個模式的button
twoS = tk.Button(win, text = " 1v1 games  ",font = ('Helvetica 18 bold',18,'bold'),bg = "white",fg = "black",anchor = "center",command = twos_window)
AiS = tk.Button(win, text = "1vsAI games",font = ('Helvetica 18 bold',18,'bold'),bg = "white",fg = "black",anchor = "center",command = AiS_window)

twoS.pack(pady = "100",anchor = "center")
AiS.pack(anchor = "center")

win.mainloop()
```
